```python
# 导入seaborn包，并设置默认绘图风格
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib as mpl

import seaborn as sns;sns.set()
```

```python
# 准备随机数据
rng = np.random.RandomState(0)
x = np.linspace(0,10,500)
y = np.cumsum(rng.randn(500,6),0)
```

```python
# 此时绘制的图像风格为sns基础风格，比matplot的默认风格好看那么一点点
plt.plot(x,y)
plt.legend('ABCDEF',frameon=False,ncol=2,loc='upper left')
```

#### 绘制频次直方图，KDE平滑分布图，两者的混合图，二维KDE图

```python
data = np.random.multivariate_normal([0,0],[[5,2],[2,2]],size=2000)
data.shape
```

```python
data = pd.DataFrame(data,columns=['x','y'])
data.head()
```

```python
# 绘制频次直方图
for col in 'xy':
    plt.hist(data[col],density=True,alpha=0.5)
```

```python
# 绘制KDE平滑分布图
for col in 'xy':
    sns.kdeplot(data[col],shade=True)
```

```python
# 同时绘制频次直方图，KDE平滑分布图
for col in 'xy':
    sns.distplot(data[col])
```

```python
# 绘制二维KDE图
g = sns.kdeplot(data['x'],data['y'])
g.axvline(x=0,linestyle='--',color='red')
```

#### 绘制联合分布图

```python
# kind='kde'
with sns.axes_style('white'): # 表示在该绘图风格上下文中进行绘图
    sns.jointplot('x','y',data,kind='kde')
```

```python
# kind='hex'
with sns.axes_style('white'):
    sns.jointplot('x','y',data,kind='hex')
```

#### 绘制分面频次直方图

```python
# 读取演示数据(小费的统计数据)
path = 'tips.csv'
tips = pd.read_csv(path)
tips.head()
```

```python
# 添加一列，显示小费占总消费额的百分比
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
tips.head()
```

```python
# 建立分类网格，对原始数据先按性别分类，再按时间段分类
grid = sns.FacetGrid(data=tips,row='sex',col='time',margin_titles=False)
# 分别统计每个分组在给定区间的数据出现频次，然后绘制直方图
grid.map(plt.hist,'tip_pct',bins=np.linspace(0,40,15))
```

```python
# 绘制带回归拟合的联合分布图,可以发现账单金额越高，给的小费越多
with sns.axes_style('white'):
    sns.jointplot('total_bill','tip',data=tips,kind='reg')
```

#### 绘制条形图

```python
# 读取有关行星发现的统计数据，进行演示
path = 'planets.csv'
planets = pd.read_csv(path)
planets.head()
```

```python
# 按照年份分类，统计每年发现的行星总数，并绘制条形图(catplot),设置每隔三年，显示一个x轴标签
with sns.axes_style('white'):
    g = sns.catplot('pl_disc',data=planets,aspect=2,color='green',kind='count')
    g.set_xticklabels(step=3)
```

```python
# 先按照年份分类，再按发现方式分类，统计每年不同方法分别发现的行星数，并绘制条形图
with sns.axes_style('white'):
    g = sns.catplot('pl_disc',data=planets,aspect=5,hue='pl_discmethod',kind='count')
    g.set_xticklabels(step=2)
```
