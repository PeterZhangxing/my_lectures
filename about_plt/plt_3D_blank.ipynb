 ### 到目前为止，我们分享了很多绘制二维图形的技巧和方法，
 ### 我详细大家都想知道如何绘制一些常用的三维图形，
 ### 下面我们就来介绍几种常用的三维图形绘制方式

```python
# 导入必要的库
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib as mpl
# 绘制3D图形必须导入该库
from mpl_toolkits import mplot3d
```

```python
# 建立3D坐标系
fig = plt.figure()
ax = plt.axes(projection='3d')
```

#### 绘制3D曲线

```python
zline = np.linspace(0,15,1000)
xline = np.sin(zline)
yline = np.cos(zline)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(xline,yline,zline,'red')
```

####  绘制3D散点图

```python
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(xdata,ydata,zdata,c=zdata,cmap='pink')
```

#### 绘制3D等高线图

```python
def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))
```

```python
x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)
# 准备网格数据
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
fig = plt.figure(figsize=(9,6))
ax = plt.axes(projection='3d')
# 绘制3D等高线图
ax.contour3D(X,Y,Z,100)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('title')
ax.set_zticks([])
```

```python
# 调整视觉角度 z,(x,y)
ax.view_init(60,30)
fig
```

#### 绘制3D线框图

```python
fig = plt.figure(figsize=(9,6))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X,Y,Z,color='pink',rstride=1,cstride=1)
```

```python
ax.view_init(60,30)
fig
```

#### 绘制3D表面图

```python
fig = plt.figure(figsize=(9,6))
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,color='pink',rstride=1,cstride=1)
```

```python
ax.view_init(60,30)
fig
```

#### 在绘制的3D散点图基础上，绘制3D表面图

```python
# 准备1000个三维坐标系中点的坐标
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = r * np.sin(theta)
y = r * np.cos(theta)
z = f(x,y)
```

```python
# 绘制的3D散点图
fig = plt.figure(figsize=(9,6))
ax = plt.axes(projection='3d')
ax.scatter3D(x,y,z,c=z,cmap='viridis',linewidth=0.3)
```

```python
ax.view_init(60,30)
fig
```

#### 绘制将离得最近的三个三维点连接起来，构成三维表面的图

```python
fig = plt.figure(figsize=(9,6))
ax = plt.axes(projection='3d')
ax.plot_trisurf(x,y,z,cmap='viridis')
```

```python
ax.view_init(60,30)
fig
```
