#### 如何从excel文件读取数据和如何将pandas中的DataFrame类型的数据保存在excel文件中

```python
import numpy as np
import pandas as pd
```

- 如何向excel文件写入DataFrame数据
    - 安装xlsxwriter作为读取引擎:pip install xlsxwriter
    - 编写代码

```python
excel_writer = pd.ExcelWriter(path='sample.xlsx',engine='xlsxwriter')
df = pd.DataFrame(np.random.randn(12,6),columns=list('ABCDEF'))
df.to_excel(excel_writer,sheet_name='第一页啊')
excel_writer.save()
```

- 如何从excel文件读取数据
    - 安装指定版本的xlrd:pip install xlrd==1.2.0
    - 编写代码

```python
df = pd.read_excel('sample.xlsx')
df
```
